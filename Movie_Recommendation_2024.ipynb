# Movie Recommendation

```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("10000_movies_dataset.csv")
```

```python
movies.head()
```

```python
movies.columns
```

```python
movies.info()
```

```python
movies['tags'] = movies['genre'] + movies['overview']
```

```python
movies.head()
```

```python
new_df = movies[['id','title','genre','overview','tags']]
```

```python
new_df = new_df.drop(columns = ['genre','overview'])
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 10000, stop_words='english')
```

```python
cv
```

```python
vec = cv.fit_transform(new_df['tags'].values.astype('U')).toarray()
```

```python
vec
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vec)
```

```python
sim
```

```python
new_df[new_df['title']=='The Shawshank Redemption']
```

```python
dist = sorted(list(enumerate(sim[0])),reverse = True,key= lambda vec:vec[1])
```

```python
dist
```

```python
for i in dist[0:5]:
  print(new_df.iloc[i[0]].title)
```

```python
def recommend(movies):
  index = new_df[new_df['title']== movies].index[0]
  distance = sorted(list(enumerate(sim[index])),reverse =True,key=lambda vec:vec[1])
  print(f"If you like {movies} You might also like:")
  count = 0
  for i in distance[1:6]:
    count+=1
    print(f"{count}.{new_df.iloc[i[0]].title}")
```

```python
recommend("The Shawshank Redemption")
```

```python
recommend("Batman")
```

```python
recommend("Superman")
```
